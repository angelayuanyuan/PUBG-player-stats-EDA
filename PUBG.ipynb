```python
fire_distance = np.sqrt((map_island['killer_position_y']-map_island['victim_position_y'])**2+(map_island['killer_position_x']-map_island['victim_position_x'])**2)
```

```python
map_island['fire_distance']=fire_distance
```

```python
map_island
```

```python
map_island.boxplot(column='fire_distance', by='killed_by',figsize=(20,10))
plt.xticks(rotation=90) # well, that's weird, you can't punch somebody miles away
```
